```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("chess_data.xlsx", sheet_name = 'Sheet2')
```

```python
df.head()
```

#### Each game is divided in two rows,i.e. The row at index 0 and 1 belong to the same game similarly 2 and 3 belong to the same game. And in chess.com data, top player always plays as white, so every player in even number of rows is a player who played as white in the game and every player in odd number of rows is a player who played as black.

#### To properly analyze the data, each game has to be 1 row, so now the purpose is to create additional columns to bring each players of the same game into the same row.

```python
#Creating new columns without any values.

df['Black Player'] = np.nan
df['Result2'] = np.nan
df['Black_Accuracy'] = np.nan
```

```python
df.head()
```

```python
#Copying data to each new columns using iloc. 

for i in range(0,2044,2):
    df.iloc[i,6] = df.iloc[i+1,1]

for i in range(0,2044,2):
    df.iloc[i,7] = df.iloc[i+1,2]

for i in range(0,2044,2):
    df.iloc[i,8] = df.iloc[i+1,3]
```

```python
df.head(10)
```

```python
#Renaming Columns with appropriate names.

df.rename(columns = {"Players":"White_Player","Result":"White_Result","Result2":"Black_Result","Accuracy":
                     "White_Accuracy"},inplace = True)
```

```python
df.head(10)
```

```python
# Dropping the rows which previously had information for black players. Since, information for each game has been collated into 
# single row, we are dropping them. 

for i in range(1,2044,2):
    df.drop([i],inplace = True)
```

```python
df.head(10)
```

```python
# Resetting the index
df.reset_index(drop = True,inplace = True)
```

```python
#Splitting the players and their ratings for White and Black. 

df[["White_Player1","White_Player_Rating"]] = df['White_Player'].str.split("(",expand = True)
df[["Black_Player1","Black_Player_Rating"]] = df['Black Player'].str.split("(",expand = True)
```

```python
df.head(3)
```

```python
#Stripping ")" from each values in Ratings    

df["White_Player_Rating"] = df["White_Player_Rating"].str.rstrip(")")
df["Black_Player_Rating"] = df["Black_Player_Rating"].str.rstrip(")")
```

```python
df.head(3)
```

```python
#Arranging the columns.

df = df[['Date','White_Player1','White_Player_Rating','White_Result','White_Accuracy','Black_Player1',
         'Black_Player_Rating','Black_Result','Black_Accuracy','Moves','Time']]
```

```python
#Renaming the columns removing "1" from the column name.

df.rename(columns = {"White_Player1":"White_Player","Black_Player1":"Black_Player"},inplace = True)
```

```python
df.head(5)
```

```python
df['White_Accuracy'].replace('Review', np.nan, inplace = True)
```

```python
df.head(3)
```

```python
#Converting the values to int 


df['White_Player_Rating'] = df['White_Player_Rating'].astype(int)
#df['White_Result'] = df['White_Result'].astype(int)
df['Black_Player_Rating'] = df['Black_Player_Rating'].astype(int)
#df['Black_Result'] = df['Black_Result'].astype(int)
df['Moves'] = df['Moves'].astype(int)
```
